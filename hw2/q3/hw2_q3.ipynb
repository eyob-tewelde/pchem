# Partition Function and Microstate Probababilities for a Spin-Up Electron in a d Orbital

```python
# Import libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define temperature range and boltzman constant

temp = np.array(list(range(10, 1010, 10)))
k = 8.617 * (10 ** -5)
```

## a) Degenerate States

```python
prob_a = [1] * len(temp)

plt.plot(temp, prob_a, label="E = 0 ev")
plt.tick_params
plt.title("Degenerate States")
plt.xlabel("Temperature (K)")
plt.ylabel("Probability")
plt.ylim(0, 1.1)
plt.xlim(10, 1000)
plt.legend(fontsize=10, loc="lower right")
plt.gca().spines[["top", "right"]].set_visible(False)
plt.savefig("hw2_q3_part_a.png")
plt.show()
```

## b) Two Groups of States (Three Lower, Two Higher)

```python
partition_b = 3 + 2 * np.exp(-0.01/(k*temp))

prob_b_low = 3 / partition_b
prob_b_high = (2 * np.exp(-0.01/(k*temp))) / partition_b
print(prob_b_high[-1])


plt.plot(temp, prob_b_low, label="E = 0 eV")
plt.plot(temp, prob_b_high, label="E = 0.01 eV")
plt.xlim(0, 1000)
plt.ylim(0, 1)
plt.xlabel("Temperature (K)")
plt.ylabel("Probability")
plt.title("Two Groups of States (Three Lower, Two Higher)")
plt.legend(fontsize=10)
plt.gca().spines[["top", "right"]].set_visible(False)
plt.savefig("hw2_q3_part_b.png")
plt.tight_layout()
plt.show()
```

## c) States with Incremental Splitting

```python
partition_c = 2 + np.exp(-0.01/(k*temp)) + np.exp(-0.03/(k*temp)) + np.exp(-0.06/(k*temp))

prob_c_1 = 2 / partition_c
prob_c_2 = (np.exp(-0.01/(k*temp))) / partition_c
prob_c_3 = (np.exp(-0.03/(k*temp))) / partition_c
prob_c_4 = (np.exp(-0.06/(k*temp))) / partition_c



plt.plot(temp, prob_c_1, label="E = 0 eV")
plt.plot(temp, prob_c_2, label="E = 0.01 eV")
plt.plot(temp, prob_c_3, label="E = 0.03 eV")
plt.plot(temp, prob_c_4, label="E = 0.06 eV")
plt.xlim(0, 1000)
plt.ylim(0, 1)
plt.xlabel("Temperature (K)")
plt.ylabel("Probability")
plt.title("States with Incremental Splitting")
plt.legend(fontsize=10)
plt.gca().spines[["top", "right"]].set_visible(False)
plt.savefig("hw2_q3_part_c.png")
plt.tight_layout()
plt.show()
```
